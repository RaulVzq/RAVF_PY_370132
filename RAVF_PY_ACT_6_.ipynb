1.- FUNCIÓN QUE LEA n CANTIDAD DE NÚMEROS HASTA QUE EL USUARIO LO DESEE, DESPLEGAR LA SUMA DE LOS NÚMEROS, MEDIA Y VALOR DE LOS NÚMEROS MAYORES Y MENORES.

```python
def N_Numeros():
    mayor=0
    menor=99999999999
    suma=0
    i=0
    resp=1
    while resp==1:
          num=int(input("INGRESA UN DIGITO: "))
          suma=suma+num
          if num > mayor:
             mayor = num
          if num < menor:
             menor = num
          resp=int(input("¿QUIERE CONTINUAR?\n SI(1) NO(2)"))
          i+=1
    print(f"LA SUMA DE TODOS LOS NUMEROS ES: {suma}")
    print(f"LA MEDIA DE LOS NUMEROS ES DE: {suma/i}")
    print(f"EL NUMERO MAYOR ES: {mayor}")
    print(f"EL NUMERO MENOR ES: {menor}")
N_Numeros()

```

2.- FUNCIÓN QUE GENERE 15 NÚMEROS IMPARES ENTRE 10 Y 60 o MÁXIMO DE 25 NÚMEROS. DESPLEGAR LA MEDIA DE LOS PARES Y MEDIA DE IMPARES.

```python
def N_Impares():
    import random
    par=0
    impar=0
    sumapar=0
    sumaimpar=0
    try:
            for i in range(0,25,1):
                num = random.randint(10,60)
                if num%2==0:
                    par=par+1
                    sumapar=sumapar+num
                    print(f"{num} ES PAR")
                elif num%2!=0:
                     impar=impar+1
                     sumaimpar=sumaimpar+num
                     print(f"{num} ES IMPAR")
                if impar==15:
                    break
            mediaimpar=sumaimpar/impar
            mediapar=sumapar/par
    except:
            print("HA OCURRIDO UN ERROR")
    else:
            if impar==15:
               print("LLEGO A LOS 15 NUMEROS IMPARES")
               print(f"FUERON: {par} NUMEROS PARES Y SUMADOS DAN: {sumapar}")
               print(f"FUERON: {impar} NUMEROS IMPARES Y SUMADOS DAN: {sumaimpar}")
               print(f"EL PROMEDIO DE LOS NUMEROS PARES ES DE: {mediapar}")
               print(f"EL PROMEDIO DE LOS NUMEROS IMPARES ES DE: {mediaimpar}")
            else:
               print("NO LLEGO A LOS 15 NUMEROS IMPARES")
               print(f"SON {impar} NUMEROS IMPARES")
               print(f"FUERON: {par} NUMEROS PARES Y SUMADOS DAN: {sumapar}")
               print(f"FUERON: {impar} NUMEROS IMPARES Y SUMADOS DAN: {sumaimpar}")
               print(f"EL PROMEDIO DE LOS NUMEROS PARES ES DE: {mediapar}")
               print(f"EL PROMEDIO DE LOS NUMEROS IMPARES ES DE: {mediaimpar}")
N_Impares()
```

3.- FUNCIÓN QUE SIRVA PARA LEER UN RANGO DADO POR EL USUARIO. REPETIR ESTA ACCIÓN HASTA QUE EL USUARIO LO DESEE, DESPLEGAR CANTIDAD DE NUMEROS Y PROMEDIO DE LOS NUMEROS..

```python
def Rango():
    resp=1
    i=0
    suma=0
    rangei=int(input("INGRESE EL RANGO INICIAL:"))
    rangefi=int(input("INGRESE EL RANGO FINAL: "))
    while resp==1:
          num=int(input(f"INGRESE UN VALOR DENTRO DEL RANGO: {rangei} Y DE: {rangefi}: "))
          if num>=rangei and num<=rangefi:
             suma=suma+num
             resp=int(input("¿DESEA CONTINUAR? SI(1) NO(2)"))
             i=i+1
             prom=suma/i
    print(f"FUERON {i} NUMEROS")
    print(f"EL PROMEDIO DE LOS NUMEROS ES DE: {prom}")
Rango()

```

4.- FUNCIÓN QUE RECIBA COMO PARÁMETRO LOS VALORES PARA EL ÁREA DE UN TRIANGULO Y RETORNE SU RESULTADO

```python
def A_Triangulo():
    try:
        base=int(input("VALOR DE LA BASE DEL TRIANGULO: "))
        if base==0:
           raise ValueError
        alt=int(input("VALOR DE LA ALTURA DEL TRIANGULO: "))
        if alt==0:
           raise ValueError
    except TypeError:
          print("DEBEN DE SER VALORES ENTEROS")
    except ValueError as e:
          print("INTRODUZCA UN DATO VALIDO")
    except:
          print("INTRODUZCA UN VALOR VALIDO")
    else:
          Area = (base * alt)/2
          print(f"EL AREA DEL TRIANGULO ES: {Area}")
A_Triangulo()
```

5.- FUNCION QUE SIRVA QUE SIRVA PARA EVALUAR EL PROMEDIO DE 3 CALIFICACIONES DADAS, SI EL USUARIO SU CALIFICACION ES APROBADA MANDAR MSGE QUE DIGA "felicidades avanzas al siguiente semestre", SI LA CALIFICACIONE ESTA REPROBADA EL MENSAGE "repetir materia " EL ALUMNO TENDRA UN MAXIMO DE CURSAR 3 VESES LA MATERIA, SI LA REPRUEBA 3 VESES MANDAR MSGE " lastima estas fuera de la UABC"

```python
def Calificaciones():
    try:
        for i in range(1,4,1):
            calif1=int(input("CALIFICACION 1ER PARCIAL: "))
            if calif1>100:
               raise ValueError
            if calif1<0:
               raise ValueError
            calif2=int(input("CALIFICIACION 2DO PARCIAL: "))
            if calif2>100:
               raise ValueError
            if calif2<0:
               raise ValueError
            calif3=int(input("CALIFICACION 3ER PARCIAL: "))
            if calif3>100:
               raise ValueError
            if calif3<0:
               raise ValueError
            prom=(calif1+calif2+calif3)/3
            if prom>=60:
               break
            elif i<3:
                print("REPETIR MATERIA")
    except ValueError as e:
         print("INTRODUZCA UN VALOR VALIDO")
    except TypeError:
         print("INTRODUZCA UN VALOR ENTERO")
    except:
         print("SUCEDIO UN ERROR")
    else:
          if prom>60:
             print("*FELICIDADES* AVANZAS AL SIGUIENTE SEMESTRE")
          if i==3:
             print("LASTIMA :(, ESTAS FUERA DE LA UABC)")
Calificaciones()
```
